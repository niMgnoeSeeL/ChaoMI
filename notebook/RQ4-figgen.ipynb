```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

#### mi-opt

```python
data_path = "../data-LocPrivacy/mi-opt.csv"
df = pd.read_csv(data_path)
joint_size = 20**2 * 340**2
df["sample ratio"] = df["num_samples"] / joint_size
# change the name of the column
df = df.rename(
    columns={"mi_emp": "Emp", "mi_miller": "Miller", "mi_chao": "ChaoSRM"}
)
df
mi_ground_truth = 3.2198044418207257
fig, ax = plt.subplots(figsize=(6, 4))
ax.axhline(
    mi_ground_truth,
    color="red",
    linestyle="--",
    label="Ground Truth",
    linewidth=6,
    alpha=0.75,
)
sns.lineplot(
    data=df,
    x="sample ratio",
    y="Emp",
    label="Emp",
    ax=ax,
    marker="o",
    markersize=8,
    linewidth=6,
    # transparency
    alpha=0.75,
)
sns.lineplot(
    data=df,
    x="sample ratio",
    y="Miller",
    label="Miller",
    ax=ax,
    marker="o",
    markersize=8,
    linewidth=6,
    alpha=0.75,
)
sns.lineplot(
    data=df,
    x="sample ratio",
    y="ChaoSRM",
    label="ChaoSRM",
    ax=ax,
    marker="o",
    markersize=8,
    linewidth=6,
    alpha=0.75,
)
ax.set_xscale("log")
plt.legend(fontsize=18)
plt.xlabel("Sample Ratio", fontsize=20)
plt.ylabel("MI", fontsize=20)
# font size
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
# tick width
plt.savefig("../figures/mi-opt-paper.pdf", bbox_inches="tight", pad_inches=0)
plt.show()
```
